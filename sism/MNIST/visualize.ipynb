```python
import torch
import matplotlib.pyplot as plt
from typing import Optional


def visualize_mnist_batch(
    tensor_batch,
    title="MNIST Images",
    with_title=False,
    save_path: Optional[str] = None,
):
    """
    Visualize a batch of MNIST images
    Args:
        tensor_batch: PyTorch tensor of shape (10, 28, 28)
        title: Title for the plot
    """
    # Convert to numpy if it's a PyTorch tensor
    if torch.is_tensor(tensor_batch):
        images = tensor_batch.detach().cpu().numpy()
    else:
        images = tensor_batch

    # Create subplot with 1 row and 10 columns
    fig, axes = plt.subplots(1, 10, figsize=(15, 2))
    fig.suptitle(title)

    for i in range(10):
        axes[i].imshow(images[i], cmap="gray")
        axes[i].axis("off")
        if with_title:
            axes[i].set_title(f"Image {i}")

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, bbox_inches="tight")
    plt.show()
```

### BBDM 

```python
dataset = torch.load("experiments/BBDM/generated_valset.pth")

for i in range(10):
    d = dataset[i]
    visualize_mnist_batch(d["img_1"][:10], title="MNIST Images - Rotated at time = 1")
    visualize_mnist_batch(d["img_0"][:10], title="MNIST Images - Ground Truth at time = 0")
    visualize_mnist_batch(d["img"][:10], title="MNIST Images - Reconstructed at time = 0 based on input at time = 1")
    print("-------" * 30)
```

### GSM

```python
dataset = torch.load("experiments/GSM/generated_valset.pth")

for i in range(10):
    d = dataset[i]
    visualize_mnist_batch(d["img_1"][:10], title="MNIST Images - Rotated at time = 1")
    visualize_mnist_batch(d["img_0"][:10], title="MNIST Images - Ground Truth at time = 0")
    visualize_mnist_batch(d["img"][:10], title="MNIST Images - Reconstructed at time = 0 based on input at time = 1")
    print("-------" * 30)
```
